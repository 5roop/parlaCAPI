# Downloading


```python
filter = {
    "filter": {
        "operator": "AND",
        "filters": [
            {"column": "text", "operator": "CONTAINS", "value": "krava"},
            {
                "operator": "OR",
                "filters": [
                    {"column": "parliament", "operator": "=", "value": "RS"},
                    {"column": "parliament", "operator": "=", "value": "SI"},
                ],
            },
        ],
    },
    # "search": {"query": "krava", "field": "text"},
    "columns": ["speaker_name", "date", "text", "cap_category"],
    "limit": 100,
    "offset": 0,
}

url = "https://parlacap.ipipan.waw.pl/"
import requests

response = requests.post(
    url + "download?format=csv",
    json=filter,
)
if not response.status_code == 200:
    raise Exception(
        f"Got weird response code: {response.status_code}, response content: {response.content}"
    )

from pathlib import Path

Path("my_downloaded_file.csv").write_bytes(response.content)
```
